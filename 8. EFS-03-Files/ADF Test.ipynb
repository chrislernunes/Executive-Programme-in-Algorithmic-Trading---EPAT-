## Read CSV file

```python
# To read csv file
import pandas as pd

# Read CSV into data frame prices_df
prices_df = pd.read_csv('AUDCAD.csv',index_col=0)
```

## Plot the price series

```python
# Import matplotlib to plot the graph
import matplotlib.pyplot as plt
%matplotlib inline

# Plot the prices series
prices_df.prices.plot(figsize=(8,4))

# Show the plot
plt.show()
```

## Perform ADF Test

```python
# To perform ADF Test
from statsmodels.tsa.stattools import adfuller

help(adfuller)
```

```python
# Compute ADF test statistics
adf = adfuller(prices_df.prices, maxlag = 1)
```

```python
adf[0]
```
